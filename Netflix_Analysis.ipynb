```python
# Check for null values
df.isnull().sum()
```

```python
# rename the listed_in to "genre"
df.rename(columns = {"listed_in" : "genre"},inplace = True)
```

```python
# Replace null values NAN to United State
df["country"].replace(np.nan , "United States",inplace = True)
```

```python
# Replace null values NAN to TV-MA
df["rating"].replace(np.nan , "TV-MA",inplace = True)
```

```python
df.isnull().sum()
```

```python
# Calculating mean of release Year
df["release_year"].value_counts().mean()
```

The Netflix Release years, visualized via HORIZONTAL BAR GRAPH.

```python
plt.title("Netflix Release Years")
df["release_year"].value_counts().plot.barh(figsize=(30,20) , color = "#EACE1D")
plt.show()
```

These are 2 graphs , one is **Bar Chart** and another one is **Pie Chart**. This is to just visulize the amount of **TV SHOWS VS MOVIES** on Netflix!.

```python
# TV SHOWS  VS MOVIES on NETFLIX - BAR CHART
plt.title("Types of Shows on Netflix")
sns.barplot(data = df , x = df["type"] , y = df.index , hue = df["type"])
plt.show()
```

The TV Show VS Movies on NETFLIX - visulized by **PIE CHART**

```python
plt.title("Type of shows on Netflix")
df["type"].value_counts().plot.pie(autopct = "%1.1f%%" ,figsize=(10,10))
plt.show()
```

The Netflix Rating Distribution, Visualized by a **COUNT PLOT**

```python
plt.figure(figsize = (15,10))
plt.title("Netflix Ratings Distribution")
sns.countplot(data = df, x = df["rating"] , hue = df["rating"])
plt.show()
```

The ratings seperated by type of release , visualised by a **COUNT PLOT**

```python
plt.figure(figsize = (15,8))
plt.title("Netflix ratings distribution separated by type of release")
sns.countplot(data = df , x = "rating" , hue = "type" , palette = ["#ACE1AF" , "#1B4D3E"])
plt.show()
```

The top 5 countries with Highest Movies/TV shows , visualised a **PIE CHART**

```python
from enum import auto
plt.figure(figsize = (15,8))
plt.title("Top 5 Netflix Company")
top_5 = df["country"].value_counts().sort_values(ascending=False).head(5)
top_5.plot.pie(autopct = "%1.1f%%",figsize = (15,8))
plt.show()
```

The Top 5 countries with highest Movies/TV Shows, visualized via **HORIZONTAL BAR GRAPH**

```python
plt.figure(figsize = (15,8))
plt.title("Top 5 Netflix Countries")
top_5 = df["country"].value_counts().sort_values(ascending=False).head(5)
top_5.plot.barh(figsize = (15,8) , color = "#E25822")
```

Top 5 Netflix distribution separated by type of release, visualizing via **BAR GRAPH**

```python
plt.figure(figsize = (15,8))
plt.title("Top 5 Netflix distribution separated by type of release")
sns.countplot(data = df , x = df["country"] , hue = "type" , order=df["country"].value_counts().iloc[:5].index , palette = ["#CD7F32" , "#E3DAC9"])
```
